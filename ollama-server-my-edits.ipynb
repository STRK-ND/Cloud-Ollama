```python
import os
import subprocess

def start_ollama_server():
    # Set environment variables
    os.environ['OLLAMA_HOST'] = "0.0.0.0"
    os.environ['OLLAMA_ORIGINS'] = "*"

    try:
        # Start the Ollama server
        print("Starting Ollama server...")
        server_process = subprocess.Popen(
            "/usr/local/bin/ollama serve",
            shell=True,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE
        )

        # Output logs to confirm the server is running
        stdout, stderr = server_process.communicate(timeout=5)  # Adjust timeout as needed
        print("STDOUT:", stdout.decode())
        print("STDERR:", stderr.decode())

    except subprocess.TimeoutExpired:
        print("The Ollama server was started successfully in the background.")
        print("Check logs for more details.")

    except Exception as e:
        print("Error starting Ollama server:", str(e))

    # Run a test command
    try:
        print("Running test command...")
        result = subprocess.run(["echo", "ollama test"], capture_output=True, text=True)
        print("Test Output:", result.stdout)
    except Exception as e:
        print("Error running test command:", str(e))

if __name__ == "__main__":
    start_ollama_server()
```
